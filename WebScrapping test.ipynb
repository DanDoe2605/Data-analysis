```python
def web_driver():
    #options = webdriver.ChromeOptions()
    options = Options()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    
    driver = webdriver.Chrome(options=options)
    return driver



def parser(a):
  
  options = Options()
  options.add_argument("--verbose")
  options.add_argument('--no-sandbox')
  options.add_argument('--headless')
  options.add_argument('--disable-gpu')
  options.add_argument("--window-size=1920, 1200")
  options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Edge(options=options)
  domain = 'https://casas.trovit.com.mx/renta-departamento-ciudad-méxico'
  driver.get(domain + '/' + str(a))
  soup = bs.BeautifulSoup(driver.page_source,'html.parser')
  return soup


```

```python
#Este ciclo scrapea página por página (23 en este caso) y guarda los objetos 'soup' en una lista
a = []
for i in range(2,25):
  soup = parser(i)
  info_dept = soup.find_all('div',{'class':'snippet-content-header'})
  a.append(info_dept)
  time.sleep(5)
```

```python
#Aquí guardamos en listas la información recogida por el ciclo parser(i)
dept = a
localizacion = []
costo = []
descripcion = []
area = []

#la sintaxis de los métodos .findAll corresponde a la etiqueta HTML seguida de la clase, id, data-qa ,etc. asociada a dicha etiqueta
for j in range (0,len(dept)):
  for i in range(0,len(dept[j])):
    descripcion.append(dept[j][i].find_all('div',{'class','item-title'})[0].text) #En el dominio elegido, la dirección y costo del inmueble
    localizacion.append(dept[j][i].find_all('span',{'class':'address'})[0].text)  #Estaban localizados bajo las etiquetas <div> y <span> correspondientes
    costo.append(dept[j][i].find_all('div',{'class':'item-price'})[0].text)       #a las clases anexas en los diccionarios adjuntos, esto se puede ver--

 # muy fácilmente entrando en el modo Inspeccionar en la página web con click derecho, desplegando el modo desarrollador
 #y  oprimiendo 'ctrl +shift+c' para localizar interactivamente la etiqueta html correspondiente al dato buscado
```

```python
#Exportamos a un dataframe
import pandas as pd
df = pd.DataFrame(list(zip(localizacion,costo,descripcion)),columns=['localizacion','costo','descripcion'])
df
```
